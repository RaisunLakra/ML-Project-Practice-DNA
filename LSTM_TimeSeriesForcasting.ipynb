# Develop LSTM Models For Time Series Forcasting

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Flatten
```

```python
def prepare_data(timeseries_data, n_features):
    X, y = [], []
    for i in range(len(timeseries_data)):
        end_ix = i + n_features
        if end_ix > len(timeseries_data) - 1:
            break
        seq_x, seq_y = timeseries_data[i:end_ix], timeseries_data[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# define input sequence
timeseries_data = [110, 125, 133, 146, 158, 172, 187, 196, 210]

# choose a no of time steps
n_steps = 3

# split into samples
X, y = prepare_data(timeseries_data, n_steps)
```

```python
print(X), print(y)
```

```python
# reshape from [samples, timesteps] into [sample, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
```

```python
print(X)
```

# Building LSTM Model

```python
# define model
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# fit model
model.fit(X, y, epochs=200, verbose=0)
```

# Predicting for next 10 data

```python
x_input = array([187, 196, 210])
temp_input = list(x_input)
lst_output = []
i = 0
while (i < 10):
    if (len(temp_input) > 3):
        x_input = array(temp_input[1:])
        print('{} day input {}'.format(i, x_input))
        # print(x_input)
        yhat = model.predict(x_input, verbose=0)
        print('{} day output {}'.format(i, yhat))
        temp_input.append(yhat[0][0])
        temp_input = temp_input[1:]
        # print(temp_input)
        lst_output.append(yhat[0][0])
    else:
        x_input = x_input.reshape((1, n_steps, n_features))
        yhat = model.predict(x_input, verbose=0)
        print(yhat)
        temp_input.append(yhat[0][0])
        lst_output.append(yhat[0][0])
    i = i + 1

print(lst_output)
```
